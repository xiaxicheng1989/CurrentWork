```python
import pandas as pd
import numpy as np
import json
```

```python
with open("Plexal_MOT_15072020.sm2") as file:
    content = file.readlines()
# everything extracted in list of strings   
content = [x.strip() for x in content] 
```

```python
# converting into one string
test = np.array(content)
newcode=""
for i in test:
    newcode+=i
```

```python
json_string = newcode
```

```python
jsonfile = json.loads(json_string)
```

Converted into json file

```python
stepsData = jsonfile['Steps']
```

```python
stepsData
```

```python
parameterData = jsonfile['Parameters']
```

```python
# Looks through the parameter data and returns a dictionary with the name and value
def getValues(parmName, pardata):
    for i in pardata:
        if i["Name"]==parmName:
            return {i["Name"] : i["Value"]}
        else:
            pass
    
```

```python
def getParameter(feature, data, parameterList):
    seqData= []
    for i in data:
        try:
            # extract parameter name for the feature channel and look for the value from the parametervaluelist
            temp = i["analogData"][feature]['_selectedItem'][1]['Value']
            seqData.append(getValues(temp,parameterList)[temp])
        except:
            # if no value, just contineou with previous value
            seqData.append(seqData[-1])
    return seqData
    
```

```python
seqNames = [i["Name"] for i in stepsData]
seqEnabled = [i["Enabled"] for i in stepsData]
# Timebase: 1 for ms 0 for us
seqDurations = [float(i["Duration"])*10**(-3) if i["Timebase"]==0 else float(i["Duration"]) for i in stepsData]
seqDigit = [i['digitalData']['digit-1'] for i in stepsData]
seqData1 = getParameter("AO-1",stepsData,parameterData) 

fileDF = pd.DataFrame({ "Seq.Name": seqNames,
                        "Seq.Enabled": seqEnabled,
                        "Seq.Duration[ms]":seqDurations,
                        "AO-1" : seqData1,
                        "digit-1":seqDigit
                      })
```
