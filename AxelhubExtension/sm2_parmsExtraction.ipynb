```python
import pandas as pd
import numpy as np
import json
```

```python
def getValues(parmName, pardata):
    for i in pardata:
        if i["Name"]==parmName:
            return {i["Name"] : i["Value"]}
        else:
            pass

def getParameter(feature, data, parameterList):
    seqData= []
    for i in data:
        try:
            # extract parameter name for the feature channel and look for the value from the parametervaluelist
            temp = i["analogData"][feature]['_selectedItem'][1]['Value']
            seqData.append(getValues(temp,parameterList)[temp])
        except:
            # if no value, just contineou with previous value
            seqData.append(seqData[-1])
    return seqData
```

```python
def getParmsDF(filename, AOlist, DIOlist):
    # getting the data in json format
    with open(filename) as file:
        content = file.readlines()
    content = ' '.join([x.strip() for x in content])
    jsonfile = json.loads(content)
    stepsData = jsonfile['Steps']
    parameterData = jsonfile['Parameters']
    
    # extract information about sequence time steps
    seqNames = [i["Name"] for i in stepsData]
    seqEnabled = [i["Enabled"] for i in stepsData]
    seqDurations = [float(i["Duration"])*10**(-3) if i["Timebase"]==0 else float(i["Duration"]) 
                    for i in stepsData]
    
    # establish dataframe
    parmsDF = pd.DataFrame({"Seq.Name": seqNames,
                            "Seq.Enabled": seqEnabled,
                            "Seq.Duration[ms]":seqDurations
                            })

    for aochannel in AOlist:
        seqData = getParameter(aochannel,stepsData,parameterData)
        parmsDF[aochannel] = seqData
    for diochannel in DIOlist:
        seqDigit = [i['digitalData'][diochannel] for i in stepsData]
        parmsDF[diochannel] = seqDigit

    return parmsDF
```

```python
AOlist = ["AO-1","AO-2","AO-0"]
DIOlist = ["digit-0","digit-1"]
getParmsDF("Plexal_MOT_15072020.sm2",AOlist, DIOlist)
```
