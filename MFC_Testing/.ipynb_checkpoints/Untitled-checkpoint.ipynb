```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
files=os.listdir("20200924")
datafile = [i for i in files if "txt" in i]
```

```python
def getData(timeStr):
    return np.array([float(i[1]) 
                      for i in list(
                          pd.read_table('20200924/'+timeStr+'_RabiFlop_Counter.txt').iloc[14:-2].iloc[:,0].index)])[-1::-1]
```

## Linear movement

```python
linTrack_No = getData('1607')
linTrack_With = getData('1617')
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(linTrack_No)
ax.plot(linTrack_With)
ax.legend(["No Compensation","With Compensation"])
ax.set_title("Linear movement: 1 m ")
ax.set_xlabel("Shots")
ax.set_ylabel("P2")

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
textstr = '\n'.join((
    r'$\Delta B_1$ = 6 mG',
    r'$\Delta B_2$ = 70 mG',
    r'$\Delta B_3$ = 40 mG',
    r'$\Delta B_4$ = -40 mG',
    r'$\Delta B_5$ = -20 mG',
    r'$\Delta B_6$ = -30 mG',))

ax.text(1.1, 0.75, textstr, transform=ax.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)
```

## Rotation

```python
rotTrack_No = getData('1635')
rotTrack_With = getData('1639')
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(rotTrack_No)
ax.plot(rotTrack_With)
ax.legend(["No Compensation","With Compensation"])
ax.set_title("180$^\circ$ rotation")
ax.set_xlabel("Shots")
ax.set_ylabel("P2")

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
textstr = '\n'.join((
    r'$\Delta B_1$ = -100 mG',
    r'$\Delta B_2$ = -50 mG',
    r'$\Delta B_3$ = 80 mG',
    r'$\Delta B_4$ = 100 mG',
    r'$\Delta B_5$ = 0 mG',
    r'$\Delta B_6$ = -80 mG',))

ax.text(1.1, 0.75, textstr, transform=ax.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)
```

## Linear and Rotation

```python
linrotlinTrack_No = getData('1623')
linrotlinTrack_With = getData('1629')
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(linrotlinTrack_No)
ax.plot(linrotlinTrack_With)
ax.legend(["No Compensation","With Compensation"])
ax.set_title("Linear movement: 0.5 m -> 90$^\circ$ rotation -> Linear movement back")
ax.set_xlabel("Shots")
ax.set_ylabel("P2")

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
textstr = '\n'.join((
    r'$\Delta B_1$ = 12 - -90 - 50 mG',
    r'$\Delta B_2$ = 50 - -1 - -50 mG',
    r'$\Delta B_3$ = 30 - 100 - -10 mG',
    r'$\Delta B_4$ = -20 - 50 - -30 mG',
    r'$\Delta B_5$ = -10 - 30 - -15 mG',
    r'$\Delta B_6$ = -10 - -90 - 40 mG',))

ax.text(1.1, 0.75, textstr, transform=ax.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)
```
