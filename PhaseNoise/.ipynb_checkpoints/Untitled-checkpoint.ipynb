```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files = [i for i in os.listdir() if "2.txt" in i]
```

```python
files
```

```python
dBmV2V = lambda dbmv: (10**(dbmv/20))/1000
V2VsqHz = lambda v, rbw: v/np.sqrt(rbw)
```

```python
def prepData(file, rbw):
    test = pd.read_table(file, names=["Freq","dBmV"])
    test["V"] =[dBmV2V(i) for i in test["dBmV"]]
    test["V/sqHz"] =[V2VsqHz(i, rbw) for i in test["V"]]
    
    test["rad"] =test["V"]*np.pi/2.5
    test["rad/sqHz"] =[V2VsqHz(i, rbw) for i in test["rad"]]
    
    return test
```

```python
rbwlist = [30000, 3000,1000, 100,10,3]
```

```python
psddata = [prepData(files[i],rbwlist[i]) for i in range(6)]
```

```python
fig, ax = plt.subplots(figsize=(8,4))
for data in psddata:
    data.plot(x="Freq", y="V/sqHz",logy=True, logx=True, ax=ax)
ax.set_ylabel("V/$\sqrt{Hz}$")
ax.legend(["10 MHz","1 MHz","100 kHz","10 kHz","1 kHz","100 Hz"])
ax.set_xlim([10,10**7])
```

## Joining data

```python
joinedPSD = psddata[-1].copy()
for df in psddata[-2::-1]:
    
    joinedPSD=joinedPSD.append(df[df["Freq"]>joinedPSD["Freq"].iloc[-1]], ignore_index=True)
    
```

```python
fig, ax = plt.subplots(figsize=(8,4))
joinedPSD.plot(x="Freq", y="V/sqHz",logy=True, logx=True, ax=ax)
ax.set_ylabel("V/$\sqrt{Hz}$")
ax.set_xlim([10,10**7])
```

## Getting numerical integration

```python
from scipy import integrate
```

```python
x = joinedPSD["Freq"].values[::-1]
y= (joinedPSD["rad/sqHz"].values[::-1])**2
y_int = -integrate.cumtrapz(y, x, initial=0)
```

```python
plt.loglog(x, np.sqrt(y_int)*1000)
plt.grid("Minor")
plt.xlabel("Offset Frequency [Hz]")
plt.ylabel("Integrated Phase noise [mrad]")
```
